seaborn tip 변수의 값(데이터) 분포, 평균과 분산

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 불러오기
tips = sns.load_dataset("tips")

# tip 변수 분포 시각화
plt.figure(figsize=(8,5))
sns.histplot(tips["tip"], kde=True)
plt.title("Distribution of Tip")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.show()

# 평균과 분산 계산
mean_tip = tips["tip"].mean()
var_tip = tips["tip"].var()

print("Mean of tip:", mean_tip)
print("Variance of tip:", var_tip)
```

평균의 함정 (Pitfall of Mean)


```python
# 평균과 중앙값 비교
median_tip = tips["tip"].median()

print("Mean:", mean_tip)
print("Median:", median_tip)

# boxplot으로 이상치 확인
plt.figure(figsize=(6,4))
sns.boxplot(x=tips["tip"])
plt.title("Boxplot of Tip (Outliers)")
plt.show()
```

3가지 표본추출방법 비교


```python
# 모집단 평균과 분산
population_mean = tips["tip"].mean()
population_var = tips["tip"].var()

print("Population Mean:", population_mean)
print("Population Variance:", population_var)

# 1. Simple Random Sampling
random_sample = tips.sample(n=50, random_state=42)

print("\nRandom Sampling")
print("Mean:", random_sample["tip"].mean())
print("Variance:", random_sample["tip"].var())


# 2. Systematic Sampling
k = 5
systematic_sample = tips.iloc[::k]

print("\nSystematic Sampling")
print("Mean:", systematic_sample["tip"].mean())
print("Variance:", systematic_sample["tip"].var())


# 3. Stratified Sampling
stratified_sample = tips.groupby("sex", group_keys=False).apply(
    lambda x: x.sample(frac=0.3, random_state=42)
)

print("\nStratified Sampling")
print("Mean:", stratified_sample["tip"].mean())
print("Variance:", stratified_sample["tip"].var())
```
